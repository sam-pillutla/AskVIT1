# MAIN TASK **CODE**

```python
# Import required libraries
import pandas as pd
import numpy as np
from faker import Faker
import random

# Initialize Faker to generate random data
faker = Faker()

# Generate a list of unique product IDs and names
product_ids = [faker.uuid4() for _ in range(20)]  # List of unique product IDs
product_names = [faker.word() for _ in range(20)]  # List of random product names

# Set the number of orders to generate
num_orders = 50

# Create a dictionary with synthetic data for orders
data = {
    'order_id': [faker.uuid4() for _ in range(num_orders)],  # Unique order IDs
    'customer_id': [faker.uuid4() for _ in range(num_orders)],  # Unique customer IDs
    'order_date': [faker.date_between(start_date='-1y', end_date='today') for _ in range(num_orders)],  # Random order dates within the last year
    'product_id': [random.choice(product_ids) for _ in range(num_orders)],  # Randomly select product IDs from the list
    'product_name': [random.choice(product_names) for _ in range(num_orders)],  # Randomly select product names from the list
    'product_price': [round(random.uniform(10.0, 100.0), 2) for _ in range(num_orders)],  # Random product prices between 10 and 100, rounded to 2 decimal places
    'quantity': [random.randint(1, 10) for _ in range(num_orders)],  # Random quantities between 1 and 10
}

# Create a DataFrame from the synthetic data
df = pd.DataFrame(data)

# Save the DataFrame to a CSV file named 'orders.csv'
df.to_csv('orders.csv', index=False)

print("Synthetic dataset generated and saved to 'orders.csv'")

# Perform analysis on the dataset
df['total_price'] = df['product_price'] * df['quantity']  # Calculate total price for each order
df['order_date'] = pd.to_datetime(df['order_date'])  # Convert order_date to datetime format
df['year_month'] = df['order_date'].dt.to_period('M')  # Extract year and month from order_date

# Calculate total revenue per month
monthly_revenue = df.groupby('year_month')['total_price'].sum().reset_index()

# Calculate total revenue per product
product_revenue = df.groupby('product_name')['total_price'].sum().reset_index()

# Calculate total revenue per customer
customer_revenue = df.groupby('customer_id')['total_price'].sum().reset_index()

# Identify the top 10 customers by total revenue
top_customers = customer_revenue.sort_values(by='total_price', ascending=False).head(10)

# Print the results of the analysis
print("Total Revenue per Month:")
print(monthly_revenue)
print("\nTotal Revenue per Product:")
print(product_revenue)
print("\nTotal Revenue per Customer:")
print(customer_revenue)
print("\nTop 10 Customers by Revenue:")
print(top_customers)
```
